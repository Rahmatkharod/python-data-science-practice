# 3.Broadcasting

```python
# Broadcasting allows NumPy to perform operations
# on arrays with different shapes automatically.
a = np.array([[2,3,4],
              [1,2,3]
             ])
b = np.array([10,20,30])   
a+b
```

```python
a = np.array([[2,3,4],
              [1,2,3]
             ])
b = np.array([10,20]).reshape(2,1)
a+b    #this doesnt work as b cant be broadcasts because its not compatible so we uses .reshape so that it will work
```

```python
#Broadcasting Example Using Ones Array
a = np.ones((3,1))
b = np.ones((1,4))
a + b
```

# 4.Boolean Masking

```python
# Boolean masking filters elements based on condition
a = np.array([10,20,30,45,30,20])
a >  30
a[a>30]
a[a%2 ==0]
```

# 5. Reshaping Arrays

```python
a = np.array([10,20,30,45,30,20])
a.reshape(-1,2)
```

# 6. Flatten vs Ravel

```python
a = np.array([[2,3,4],[1,2,3]])
b = a.flatten()
b[1] = 100
a  #flatten only copies not record changes
```

```python
a = np.array([[2,3,4],[1,2,3]])
b = a.ravel()
b[1] = 100
a   #ravel is a view...memory efficient, record changes
```

# 7. Stacking Arrays

```python

a = np.array([1,2,3])
b = np.array([4,5,6])
np.vstack((a,b,))
```

```python
np.hstack((a,b))
```

```python
np.concatenate((a,b))  #same as hstack
```
